# Comparison between SARSA and Q-Learning Agents

In this section, we compare the performance of the tuned versions of SARSA and Q-Learning agents. The comparison focuses on evaluating the two agents over a specific training setup.

#### **Evaluation Setup**
- Each evaluation consists of **1,000 iterations**.
- During each iteration, both agents play **10 episodes**.
- The reward for each iteration is calculated as the **average reward across the 10 episodes**.

#### **Results**
The final evaluation results are saved in a JSON file named comparison_model.json. This file includes the mean reward and standard deviation for each agent, which allows for further analysis and reproducibility of the experiment.


## Libraries

```python
import gymnasium as gym
from agents.SarsaAgent import SarsaAgent
from agents.QLearningAgent import QLearningAgent
from utils.tile_coding import mc_tile_encoding
from utils.utils_fct import compare_with_moving_averages, compare_agents, process_json
```

## Comparison on evaluation

```python
# Set up the agents 
sarsa_agent = SarsaAgent(env=gym.make("MountainCar-v0"), encode_fct=mc_tile_encoding)
qlearning_agent = QLearningAgent(env=gym.make("MountainCar-v0"), encode_fct=mc_tile_encoding)


# Train the agents
sarsa_agent.train(nb_episodes=1000, alpha=0.1, gamma=0.99, epsilon=0.01, use_glei=True)
qlearning_agent.train(nb_episodes=1000, alpha=0.1, gamma=0.99, epsilon=0.01, use_glei=True)
```

```python
compare_with_moving_averages(
    agent1_dict= { "name" : "SARSA", "agent" : sarsa_agent},
    agent2_dict= { "name" : "Q-Learning", "agent" : qlearning_agent},
    num_episodes=1000,
    window_size=100
)
```

```python
compare_with_moving_averages(
    agent1_dict= { "name" : "SARSA", "agent" : sarsa_agent},
    agent2_dict= { "name" : "Q-Learning", "agent" : qlearning_agent},
    num_episodes=1000,
    window_size=100
)
```

```python
means, stds = compare_agents(
    agent1_dict= { "name" : "SARSA", "agent" : sarsa_agent},
    agent2_dict= { "name" : "Q-Learning", "agent" : qlearning_agent},
    num_episodes=1000,
    verbose=True
)
```

```python
process_json("./data/comparison_model", "w", data_dict={
    "SARSA" : {"mean" : means[0], "std" : stds[0]}, 
    "Q-Learning" : {"mean" : means[1], "std" : stds[1]}
})
```
